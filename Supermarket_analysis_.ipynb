```python
import numpy as np
import pandas as pd

sup = pd.read_csv('/content/supermarket_sales.csv')
sup.head()
```

```python
sup.info()
```

```python
sup['Date'] = pd.to_datetime(sup['Date'])
sup.info()
```

```python
sup['Time'] = pd.to_datetime(sup['Time'], format = '%H:%M').dt.time
sup.info()
```

```python
sup.head(10)
```

```python
#round up the float values to 2 decimal places
round_cols = ['Unit price', 'Tax 5%', 'Total', 'cogs',	'gross margin percentage',	'gross income']
for col in round_cols:
 sup[col] = sup[col].round(2)

sup.head()
```

```python
#REARRANGE THE COLUMNS FOR EASY READABILITY

# Define the NEW column order
new_order = ['Invoice ID', 'Branch', 'City', 'Customer type', 'Gender',
             'Month', 'Date', 'Time', 'Payment', 'Product line',
             'Quantity', 'Unit price', 'cogs', 'Tax 5%', 'Total',
             'gross margin percentage', 'gross income', 'Rating']

# Rearrange columns in the DataFrame
sup = sup[new_order]

sup.head()
```

```python
#Sales and Revenue Analysis Questions:
#Which product line generates the most revenue?
#What is the average sales amount per invoice?
#How do sales vary by month, date, and time?
#Which city has the highest sales and profitability?
```

### QUESTION 12 CLASSWORK

```python
#Which product line generates the most revenue?
sup.groupby('Product line')['Total'].sum().sort_values(ascending = False)

```

```python
#What is the average sales amount per invoice?
avg_salper_inv = sup['Total'].sum()/sup['Invoice ID'].nunique()
print(f'The average sales per invoice is: ${avg_salper_inv:.2f}')
```

```python
#How do sales vary by month, date, and time?

monthly_sales = sup.groupby('Month')['Total'].sum().sort_values(ascending = False)
monthly_sales
```

```python
hourly_sales = sup.groupby('Time')['Total'].sum().sort_values(ascending = False)
hourly_sales
```

```python
sup.groupby(['Time', 'Product line'])['Total'].sum().sort_values(ascending = False)
```

```python
daily_sales = sup.groupby('Date')['Total'].sum().sort_values(ascending = False)
daily_sales
```

```python
#Profitability:

#Which product line has the highest profit margin?
#What is the average gross margin percentage across branches and product lines?
#Are there branches or product lines with lower profitability that need attention?
```

```python
sup.head()
```

```python
sup.groupby('Product line')['gross income'].sum().sort_values(ascending = False)
```

```python
#What is the average gross margin percentage across branches and product lines?
sales = sup.groupby(['City','Product line'])['gross income'].mean().unstack()
sales
```

```python
import matplotlib.pyplot as plt

sales.plot(kind = 'bar', figsize = (10,6))
plt.xlabel('City')
plt.ylabel('Gross Income')
plt.title('Gross Income by City and Product Line')
plt.legend(title = 'Product line', bbox_to_anchor=(1.05, 1), loc='upper left') # Moved legend outside
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

```python
#Are there branches or product lines with lower profitability that need attention?

profitability= sup.groupby(['City', 'Product line'])['gross income'].sum().sort_values()
profitability

```

```python
#product line with the least sales and why?
pdl_sales = sup.groupby('Product line')['Total'].sum().sort_values(ascending = False)
least_sales = pdl_sales.index[-1]
least_sales_values = pdl_sales.iloc[-1]
print(f'The product line with the least sales is {least_sales} with a total of ${least_sales_values:.2f}')
```

Possible reasons for low sales
1. High competition: Competitors might offer similar products at lower prices or with better features.
2. Poor marketing: Insufficient marketing efforts could lead to low customer awareness and interest.
